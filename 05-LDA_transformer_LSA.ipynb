```python
def summarize_article(article, vectorizer, lsa, num_sentences=5): # five sentences
    sentences_tokens = article.split('. ') # manual sentence tokens the article into sentences - split on fullstop
    #vectorizer_tfidf = TfidfVectorizer(stop_words=news_stop_words) #tf-idf vectorizer- custom stopwords
    vectorized_text = vectorizer_tfidf.fit_transform(sentences_tokens)  # sentences into vectors - numerical representations
    #lsa = TruncatedSVD(n_components=2) # lsa -load -dimensionality reduction
    lsa_matrix = lsa.fit_transform(vectorized_text) # apply lsa to vectors
    total_scores_summarised = lsa_matrix.sum(axis=1) # sum of scores for each sentence
    rank = np.argsort(total_scores_summarised)[::-1]  # indices of sentences by importance
    top_sentences = [sentences_tokens[i] for i in rank[:num_sentences]]
    return '. '.join(top_sentences) # join sentences together again
```

```python
dataset_bart_train['generated_summary_testing'] = dataset_bart_train['article'].apply(
    lambda x: summarize_article(x, vectorizer_tfidf, lsa))
```

```python
import pandas as pd
from transformers import pipeline
import torch

def summarise_docs(doc, model_name='t5-small', max_output_length=50, min_output_length=50): # 50 tokens

    device = 0 if torch.cuda.is_available() else -1 # GPU if available - faster parrallel processing - speed
    summarizer = pipeline("summarization", model=model_name, tokenizer=model_name, device=device)

    # summarise each doc
    return [summarizer(doc, max_length=max_output_length, min_length=min_output_length, do_sample=False)[0]['summary_text'] for doc in doc]

def add_df(df, text_column, model_name='t5-small', max_output_length=50, min_output_length=50):
    df['transformer_summary'] = summarise_docs(df[text_column], model_name, max_output_length, min_output_length)
    return df


transformer_summaries = add_df(dataset_bart_train, 'generated_summary_testing')
print(transformer_summaries)
```

```python
import nltk
from nltk.tokenize import word_tokenize
from gensim import corpora
```

```python
# load english stopwords - NLTK Library
stop_words = stopwords.words('english')

print('English stop words:', stop_words)

print('Punctuation from string lib: ', string.punctuation)
```

```python
from itertools import filterfalse # filter the stopwords out
```

```python
def lda_tok(doc, stop_words):
    word_tokens_lda = word_tokenize(doc)
    filtered_word_tokens_lda = list(filterfalse(stop_words.__contains__, word_tokens_lda))
    cleaned_tokens = [re.sub('[,.]', '', token) for token in filtered_word_tokens_lda]
    return cleaned_tokens
```

```python
# apply the tokenizer function
processed_lda = [lda_tok(doc, stop_words) for doc in transformer_summaries['transformer_summary']]
processed_highlights = [lda_tok(doc, stop_words) for doc in transformer_summaries['highlights']]
```

```python
# dictionary
dictionary = corpora.Dictionary(processed_lda)
dictionary_highlights = corpora.Dictionary(processed_highlights)
#  bag-of-words
diction = [dictionary.doc2bow(doc) for doc in processed_lda]
diction_highlights = [dictionary.doc2bow(doc) for doc in processed_highlights]
```

```python
from gensim.models import LdaModel

# train  model - lda
lda_model = LdaModel(diction, num_topics=2, id2word=dictionary, passes=3)
lda_model_highlights = LdaModel(diction_highlights, num_topics=2, id2word=dictionary_highlights, passes=3)
```
