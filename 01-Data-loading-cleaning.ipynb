**APPLYING CLEANING TO TESTING AND VALIDATION DATA**

```python
# going over the training data again with the function to ensure consistency with test and validation. 
```

```python
# Load the data

vdata = pd.read_csv('/Users/oyindaomole/Documents/NLP/Data/validation.csv')
tdata = pd.read_csv('/Users/oyindaomole/Documents/NLP/Data/test.csv')
trdata = pd.read_csv('/Users/oyindaomole/Documents/NLP/Data/train.csv')
```

```python
validation_data = vdata.copy() # work with copy to prevent data leakage
test_data = tdata.copy() # work with copy to prevent data leakage
train_data = trdata.copy() # work with copy to prevent data leakage
```

```python
import re

def clean_text(text):
    # Strip whitespace
    text = text.strip()
    #remove ./n and -- and - (newline characters ect custom specific)
    text = re.sub(r'\.\n|--|-', ' ', text)
    #remove meta data
    text = re.sub(r'^By \. [A-Za-z\s]+ \. PUBLISHED: \. \d{2}:\d{2} [A-Z]{3}, \d{1,2} [A-Za-z]+ \d{4} \. \| \. UPDATED: \. \d{2}:\d{2} [A-Z]{3}, \d{1,2} [A-Za-z]+ \d{4} \.?'
, ' ', text)
    # Remove bracketed information
    text = re.sub(r'[\[\(](.*?)[\]\)]', ' ', text)
    # Remove HTML tags
    text = re.sub(r'<[^>]+>', '', text)
    # Remove special characters 
    text = re.sub(r'[^\w\s.,]', '', text)
    # Lowercase the text - consistency and normalisation
    text = text.lower()
     # Remove extra whitespace
    text = re.sub(r'\s+', ' ', text) # put at the end of text after all trasformations
    return text
```

```python
# cleaning all datasets - separately to columns - they have different types of info

validation_data['article'] = validation_data['article'].apply(clean_text)
test_data['article'] = test_data['article'].apply(clean_text)

validation_data['highlights'] = validation_data['highlights'].apply(clean_text)
test_data['highlights'] = test_data['highlights'].apply(clean_text)
```

```python

train_data['article'] = train_data['article'].apply(clean_text)
train_data['highlights'] = train_data['highlights'].apply(clean_text)
```
